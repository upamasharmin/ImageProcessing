```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

We know there are five steps to a continuous wavelet transform.
1. Take a wavelet & compare it to a section at the start of
the original signal.
2. Calculate a correlation coefficient c
3. Shift the wavelet to the right & repeat steps 1 & 2 until I have
covered the whole signal.
4.Scale (stretch) the wavelet & repeat steps 1 through 3.
5. Repeat steps 1 through 4 for all scales.

```python
t = np.linspace(0,1,200)
#Finding signal by adding three different signals
signal = np.cos(2 * np.pi * 7 * t) + np.real(np.exp(-7*(t-0.4)**2)*np.exp(1j*2*np.pi*2*(t-0.4)))
#No .of scales
scales = np.arange(1,31)
#Finding CWT using Gaussian wavelet
coef,freqs =pywt.cwt(signal, scales, 'gaus1')
```

Scaling lets you either narrow down the frequency band of interest, or
determine the frequency content in a narrower time interval.
Low scale:a Compressed wavelet Rapidly changing detailsHigh
frequency
High scale: a Stretched wavelet  Slowly changing, coarse features
 Low frequency


```python
#Plotting scalogram
plt.figure(figsize=(15,20));
plt.imshow(abs(coef),extent=[0,200,30,1],interpolation='bilinear',cmap ='bone',aspect ='auto',vmax =abs(coef).max(),vmin=abs(coef).min())
plt.gca().invert_yaxis()
plt.yticks(np.arange(1,31,1))
plt.xticks(np.arange(0,201,10))
plt.show()
```

The scale factor works exactly the same with wavelets.
The smaller the scale factor, the more "compressed"
the wavelet.
